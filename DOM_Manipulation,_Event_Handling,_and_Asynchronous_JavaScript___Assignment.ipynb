#Question 2 : You need to change the appearance of a button when users hover over it. Write code that finds a button with id "myButton" and changes its background color to blue and text color to white using the element.style property.

```
<script>
    var btn = document.getElementById("myButton");

    btn.onmouseover = function () {
        btn.style.backgroundColor = "blue";
        btn.style.color = "white";
    };

    btn.onmouseout = function () {
        btn.style.backgroundColor = "";
        btn.style.color = "";
    };
</script>
```

First, the getElementById() method is used to find the button whose id is "myButton".

When the user moves the mouse over the button, the onmouseover event occurs. At that time, the button’s background color is changed to blue and the text color is changed to white using the style property.

When the mouse moves away from the button, the onmouseout event restores the button to its original appearance.


#Question 3 : Create a button that shows an alert with the message "Button was clicked!" when clicked. Also, add a second event listener to the same button that changes its text to "Clicked!" after being clicked. Explain why addEventListener is better than using onclick.

```
<button id="myButton">Click Me</button>

<script>
    var btn = document.getElementById("myButton");

    // First event: show alert
    btn.addEventListener("click", function() {
        alert("Button was clicked!");
    });

    // Second event: change button text
    btn.addEventListener("click", function() {
        btn.innerText = "Clicked!";
    });
</script>
```
- addEventListener() is used to attach one or more events to an element.

- Here, the first click event shows an alert message, and the second click event changes the button text.

- Unlike onclick, addEventListener() allows multiple event handlers on the same element without overwriting each other.


# Question 4 : Write a function named extractDomain that takes an email string like "john@example.com" and returns the domain part ("example.com") using only string methods. Do not use regular expressions.

```
function extractDomain(email) {
    var atPosition = email.indexOf("@"); // find position of "@"
    var domain = email.slice(atPosition + 1); // get substring after "@"
    return domain;
}

// Example:
console.log(extractDomain("john@example.com")); // Output: example.com
```


# Question 5 : Write a function called repeatMessage that takes a message string and a number, and returns the message repeated that many times, separated by a space. Do not use loops.
#Example:
#repeatMessage("hello", 3) should return "hello hello hello"



```
function repeatMessage(message, times) {
    return Array(times).fill(message).join(" ");
}

// Example:
console.log(repeatMessage("hello", 3)); // Output: "hello hello hello"
```

# Question 6 : Explain what the Event Loop is in JavaScript and predict the output of the following code. Also explain why the output appears in that specific order.
```
console.log('Start');
setTimeout(function() {
console.log('Timeout 1');
}, 0);
console.log('Middle');
setTimeout(function() {
console.log('Timeout 2');
}, 0);
console.log('End');
```

- Explanation of Event Loop:

The Event Loop in JavaScript is a mechanism that allows JavaScript to perform non-blocking operations even though it is single-threaded.

It keeps checking the call stack and the task queue. If the call stack is empty, it takes the first task from the queue and executes it.

Functions inside setTimeout go to the task queue after the specified time, even if the delay is 0. They wait until the call stack is empty.

- Output:-
```
Start
Middle
End
Timeout 1
Timeout 2
```

1. 'Start' is logged immediately because it is on the call stack.

2. setTimeout(..., 0) schedules Timeout 1 to the task queue, but it does not run immediately.

3. 'Middle' is logged next because it is the next statement in the call stack.

4. The second setTimeout(..., 0) schedules Timeout 2 in the task queue.

5. 'End' is logged.

6. Now the call stack is empty, so the event loop takes the first task from the queue (Timeout 1) and logs it.

7. Then it takes the next task (Timeout 2) and logs it.



# Question 7 : Write a JavaScript function named startCountdown that logs “Time left: X” every second, starting from 5 down to 1, and finally logs “Time’s up!”. Use setInterval and clearInterval only.

```
function startCountdown() {
    let time = 5;

    let intervalId = setInterval(function() {
        console.log("Time left: " + time);
        time--;

        if (time === 0) {
            console.log("Time's up!");
            clearInterval(intervalId); // Stop the interval
        }
    }, 1000);
}

// Call the function
startCountdown();
```


# Question 8 : Explain what Promises are in JavaScript and write code that creates a Promise to simulate checking if a user is logged in. The Promise should resolve with user data after 2 seconds if successful, or reject with an error message if the user is not found.

- Explanation of Promises:

A Promise in JavaScript is an object that represents the result of an asynchronous operation.

It can be in one of three states:

Pending - The operation is not finished yet.

Resolved (Fulfilled) - The operation completed successfully.

Rejected - The operation failed.

Promises are useful to handle asynchronous tasks without deeply nested callbacks.

```
function checkUserLogin(userExists) {
    return new Promise(function(resolve, reject) {
        setTimeout(function() {
            if (userExists) {
                resolve({ name: "John", email: "john@example.com" });
            } else {
                reject("User not found");
            }
        }, 2000); // 2 seconds delay
    });
}

// Example usage:
checkUserLogin(true)
    .then(function(userData) {
        console.log("User logged in:", userData);
    })
    .catch(function(error) {
        console.log("Error:", error);
    });
```

- Rewritten Code Using Async/Await:
```
async function getUserDataAndPosts(userId) {
    try {
        const userData = await fetchUserData(userId); // wait for user data
        const posts = await fetchUserPosts(userData.id); // wait for user's posts
        return { user: userData, posts: posts };
    } catch (error) {
        console.error("Error:", error);
        throw error; // re-throw the error
    }
}

// Example usage:
getUserDataAndPosts(1)
    .then(result => console.log(result))
    .catch(error => console.log("Caught error:", error));
```

- JavaScript Code Using Fetch API:

```
const apiKey = 'abcd1234';
const apiUrl = `https://api.openweathermap.org/data/2.5/weather?q=Mumbai&appid=${apiKey}&units=metric`;

fetch(apiUrl)
    .then(response => {
        if (!response.ok) {
            throw new Error("Network response was not ok");
        }
        return response.json();
    })
    .then(data => {
        // Display weather information
        console.log(`City: ${data.name}`);
        console.log(`Temperature: ${data.main.temp}°C`);
        console.log(`Weather: ${data.weather[0].description}`);
        console.log(`Humidity: ${data.main.humidity}%`);
    })
    .catch(error => {
        console.error("Error fetching weather data:", error);
    });
```
